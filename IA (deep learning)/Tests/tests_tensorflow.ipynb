```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

## Calcul dérivées première et seconde d'une fonction f(x) 

```python
# Points de départ et d'arrivée + nombre de points
x_i = -100
x_f = 100
nsteps = 100

# Array pour tracer les résultats finaux
x_range = np.linspace(x_i, x_f, nsteps)

# Tenseur utilisé pour calculer les gradients
x = tf.Variable(initial_value = np.linspace(x_i, x_f, nsteps))
```

## Calcul des gradients imbriqués

```python
# Dérivée seconde
with tf.GradientTape() as second_order_gradient:
    # Dérivée première
    with tf.GradientTape() as first_order_gradient:
        y = 2*x**2 + 3*x + 8
    first_derivative = first_order_gradient.gradient(y, x)
second_derivative = second_order_gradient.gradient(first_derivative, x)
```

## Résultats analytiques pour comparaison

```python
def f(x):
    return 2*x**2 + 3*x + 8
```

```python
def f_prime(x):
    return 4*x + 3
```

```python
def f_prime2(x):
    return 4
```

## Tracé des gradients obtenus et des expressions analytiques

```python
verif = [f(i) for i in x]

#plt.plot(x_range, verif)
plt.plot(x_range, y, x_range, verif)
plt.show()
```

```python
verif1 = [f_prime(i) for i in x]

#plt.plot(x_range, verif1)
plt.plot(x_range, first_derivative, x_range, verif1)
plt.show()
```

```python
verif2 = [f_prime2(i) for i in x]

#plt.plot(x_range, verif2)
plt.plot(x_range, second_derivative, x_range, verif2)
plt.show()
```
