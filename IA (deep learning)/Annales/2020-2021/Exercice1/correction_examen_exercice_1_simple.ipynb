```python
import numpy as np
import matplotlib.pyplot as plt    

data = np.genfromtxt("data_1D_examen.csv", delimiter=",")
nRow, nCol = data.shape
print(f"There are {nRow} rows and {nCol} columns")
```

```python
nbx_train = 175

x_train = data[:nbx_train, 0]
y_train = data[:nbx_train, 1]

x_test = data[nbx_train:, 0]
y_test = data[nbx_train:, 1]
nbx_test = np.size(x_test)
```

```python
plt.plot(x_train, y_train,".")
plt.show()
```

```python
plt.plot(x_test, y_test,".")
plt.show()
```

```python
train_data = x_train.reshape((nbx_train,1))
train_targets = y_train
test_data = x_test.reshape((nbx_test,1))
test_targets = y_test
```

**Model definition**

```python
from tensorflow import keras
from tensorflow.keras import layers

def build_model():
    model = keras.Sequential([
        layers.Dense(512, activation="relu"),
        layers.Dense(512, activation="relu"),
        layers.Dense(1)
    ])
    model.compile(optimizer="rmsprop", loss="mse", metrics=["mae"])
    return model
```

```python
model = build_model()
history = model.fit(train_data,
                    train_targets,
                    epochs=800,
                    batch_size=16,
                    validation_data=(test_data, test_targets)) 
```

```python
history_dict = history.history
history_dict.keys()
```

```python
history_dict = history.history
loss_values = history_dict["loss"]
val_loss_values = history_dict["val_loss"]
epochs = range(1, len(loss_values) + 1)
plt.plot(epochs, loss_values, "r.", label="Training loss")
plt.plot(epochs, val_loss_values, "b", label="Validation loss")
plt.title("Training and validation loss")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend()
plt.show()
```

```python
debut_epochs = 100
epochs_restriction = range(debut_epochs+1, len(loss_values) + 1)
plt.plot(epochs_restriction, loss_values[debut_epochs:], "r.", label="Training loss")
plt.plot(epochs_restriction, val_loss_values[debut_epochs:], "b", label="Validation loss")
plt.title("Training and validation loss")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend()
plt.show()
```

```python
plt.clf()
mae = history_dict["mae"]
val_mae = history_dict["val_mae"]
plt.plot(epochs, mae, "r.", label="Training MAE")
plt.plot(epochs, val_mae, "b", label="Validation MAE")
plt.title("Training and validation MAE")
plt.xlabel("Epochs")
plt.ylabel("MAE")
plt.legend()
plt.show()
```

```python
plt.clf()
truncation = 10
plt.plot(epochs[truncation:], val_mae[truncation:], "b", label="Validation MAE")
plt.title("Validation MAE")
plt.xlabel("Epochs")
plt.ylabel("MAE")
plt.legend()
plt.show()
```

**Training the final model**

```python
model = build_model()
model.fit(train_data, train_targets,
          epochs=500, batch_size=16, verbose=0)
test_mse_score, test_mae_score = model.evaluate(test_data, test_targets)
```
